```python
# Import neccessary libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data
df_train = pd.read_csv(
    './data/train.csv',
    #dtype={'Field_34': float}
)
df_test = pd.read_csv('./data/test.csv')

# df_train.iloc[43, 35] and df_test.iloc[42, 34] should have type float, but having str instead
```

```python
# Get the number of samples
num_samples = len(list(df_train.index))
num_samples
```

```python
# Observe the missing values
null_counts = pd.DataFrame(df_train.isnull().sum(axis = 0))
null_counts.reset_index(inplace = True)
null_counts.columns = ['Field', 'Number of null']
null_counts['% null'] = null_counts['Number of null'] / num_samples
null_counts.sort_values(
    ['Number of null'], 
    ascending = False,
    axis = 0,
    inplace = True
)
null_counts
```

```python
# Bad fields are fields that lack more than 70% value
bad_fields = null_counts['Field'][null_counts['% null'] > 0.7]
print(bad_fields)
print(bad_fields.values)
```

```python
# Drop all bad fields
df_train.drop(
    list(bad_fields),
    axis = 1,
    inplace = True
)
df_train
```

```python
# Drop the id column
df_train.drop(
    ['id'],
    axis = 1,
    inplace = True
)
```

```python
list_fields = []
list_unique_val = []
for col in df_train.columns:
    list_fields.append(col)
    list_unique_val.append(len(df_train[col].unique()))
count_unique = pd.DataFrame(
    data = {
        'Field': list_fields,
        'Num of unique': list_unique_val
    }
)
count_unique
```
